```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("airbnb_dataset.csv")
```

```python
df.head(5)
```

```python
df.columns
```

```python
# checkin missing value
print(df.isnull().sum())
```

```python
df.info()
```

```python
#Handling missing values
#coerce - change nan data to not a thing(nat)
#1. change dataset of last review from object to date time format
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
#df.head(5)
df.info()
```

```python
df.fillna({'reviews per month' : 0,'last review':df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=['license','house_rules'], errors='ignore')
```

```python
df.head(5)
```

```python
# Remove dollar signs and convert to float
df['price'] = df['price'].replace('[\$,]','', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head(5)
```

```python
# remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
#descriptive statistics
df.describe()
```

# visualization


1. What is the distribution of listing prices?


```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50,kde= True, color = 'red')
plt.title('Distribution of Listing Price')
plt.xlabel('Price $')
plt.ylabel('frequency')
plt.show()
```

2. How are Different Room Types Distributed ?

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x= 'room type', data= df, color='hotpink' )
plt.title('Room type Distribution')
plt.xlabel('room type')
plt.ylabel('count')
plt.show()
```

3. How are listings distributed across different neighbourhoods?
   

```python
plt.figure(figsize=(12,8))
sns.countplot(y='neighbourhood group', data = df, color='lightgreen', order = df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by Neighbourhood Group')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

4. What is the Relationship Between Price and RoomType?
   

```python
plt.figure(figsize=(10,6))
sns.boxplot(x= 'room type', y='price', hue='room type', data=df,palette='Set1')
plt.title('Price Vs Room type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```

5. How has the Number of reviews change over time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
